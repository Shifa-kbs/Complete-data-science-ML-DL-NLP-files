### Pandas-Dataframe and series

Pandas is a powerful data manipulation libraray in python widely used for data analysis and data cleaning.It provides two primary data strucures : Series and Dataframes.A series is a one-dimensional-array like object,while a Dataframe is a two-dimensional,size-mutable,and potentially heterogeneous tabular data structure with labeled axes(rows and columns)

```python
import pandas as pd
```

```python
## Series
# A Pandas Series is a one-dimensional array-like object that can hold any data type.It is similar to a column in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series :\n",series)
print(type(series))
# here in output first column is a index
```

```python
## Create a series from dictionary
data={'a':1,'b':2,'c':3}
series_dic=pd.Series(data)
print(series_dic)
# a,b,c are index
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## Create a dataframe from a dictionary of list

data={
    'Name':['Shifa','John','Jack'],
    'Age':[20,30,45],
    'City':['Bangalore','New york','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a dataframe from a list of dictionary

data=[
    {'Name':'Shifa','Age':20,'City':'Bangalore'},
    {'Name':'Jack','Age':30,'City':'New york'},
    {'Name':'John','Age':45,'City':'Florida'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
