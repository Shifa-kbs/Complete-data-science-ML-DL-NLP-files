### Data Visualization with Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high level interface for drawing attractive and informative statistical graphics.Seaborn helps in creating complex visualization with just a few lines of code.

```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn
tips=sns.load_dataset('tips')
tips # tips is a built-in demo dataset that comes with Seaborn
```

```python
# Create a Scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips) # 3rd parameter data specifically used in sns scatter plot
plt.title('Scatter plot of total bill vs tip')
plt.show()
```

```python
# Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total bill by size')
plt.show()
```

```python
## Categorical lots
# BAR plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total bill by day")
plt.show()
```

```python
# Box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

Bottom Whisker  > Approx. 5th percentile (or min without outliers)
Bottom of Box >	25th percentile (Q1)
Line in Middle	> 50th percentile (Median)
Top of Box	> 75th percentile (Q3)
Top Whisker	> Approx. 95th percentile (or max without outliers)
Dots above box > 	Outliers (values beyond ~1.5 IQR from Q1/Q3)

```python
# Violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
#KDE=Kernel density estimate is a smooth curve that shows where the data is concentrated (peaks) and where it's sparse (valleys).
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
sns.pairplot(tips)
```

pairplot() does not include categorical variables as axes â€” it only works with numeric (continuous or discrete) data for plotting.



```python
# Heatmap related to correlation a special functionality in seaborn
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# Plot total sales by product
plt.figure(figsize=(9,5))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
# x and y These tell Seaborn which columns to use for the axes.
plt.title('Total sales by product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
# x label Sets a clean, custom label for x-axis and y label for y axis
plt.show()
```

```python
# Plot total sales by product
plt.figure(figsize=(9,5))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
# x and y These tell Seaborn which columns to use for the axes.
plt.title('Total sales by region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
# x label Sets a clean, custom label for x-axis and y label for y axis
plt.show()
```
